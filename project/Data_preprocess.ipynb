# Data preprocess
This script is for data preprocessing 
## 1 read the txt file and save as a csv

```python
import pandas as pd
import os
import re
```

```python
df=pd.DataFrame(columns=['生日','人物','t_start','t_end','地点','机构','职位'])
birthday_df = pd.DataFrame(columns=['生日','人物'])
link_list=["－","—","-"]
for txt_file in sorted(os.listdir('./SZ_data')):
    with open('./SZ_data/'+txt_file,'rt') as f:
        lines = f.readlines()
        base_dict = {}
        print(txt_file)
        for line in lines:
            if line !='\n':
                #print(line)
                ogz_idx = line.index("机构") # organizaiton
                pos_idx = line.index("职位") # position
                ogz_str = ';'.join(line[ogz_idx:pos_idx].split())
                pos_str = ';'.join(line[pos_idx:].split())

                temp_dict = base_dict.copy()
                line_list = line[:ogz_idx].split()
                line_list.append(ogz_str)
                line_list.append(pos_str)
                for field in line_list:
                    kv_pair = field.split("：")
                    if len(kv_pair)!=2:
                        raise Exception("cannot split key-value: ",kv_pair)
                    k,v = kv_pair
                    # print(k,v)
                    if k=="时间" and "人物" not in line:
                        splited_time = re.split("—|-|－",v)
                        if len(splited_time)!=2:
                            raise Exception("cannot split time: ",splited_time)
                        start,end = splited_time
                        temp_dict['t_start']=start
                        temp_dict['t_end']=end
                    else:
                        temp_dict[k]=v
#                 print(line,end="")
#                 print(temp_dict)
                if "人物" in line:
                    base_dict["生日"]=temp_dict["时间"]
                    base_dict["人物"]=temp_dict["人物"]
                    birthday_df = birthday_df.append(base_dict,ignore_index=True)
#                     print(base_dict)
                else:
#                     print('appending')
                    df = df.append(temp_dict,ignore_index=True)
    f.close()
df.to_csv('./raw_data.csv',index=False)
birthday_df.to_csv('./birthday.csv',index=False)
```
