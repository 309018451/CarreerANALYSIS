```python
# -*- coding: utf-8 -*-
"""
Created on Thu Apr  9 10:55:22 2020

@author: ASUS
"""
import sklearn
import numpy as np
import re
import random
from gensim.models import Word2Vec
import networkx as nx
import sys
from multiprocessing import cpu_count
import time
import argparse
import os
import numpy as np
from munkres import Munkres
from sklearn import datasets

import matplotlib.pyplot as plt
from pylab import *
mpl.rcParams['font.sans-serif'] = ['SimHei']


from gensim.models import KeyedVectors
#from utils.similarity import calEuclidDistanceMatrix
#from utils.knn import myKNN
#from utils.laplacian import calLaplacianMatrix
#from utils.dataloader import genTwoCircles
#from utils.ploter import plot
#from sklearn.cluster import KMeans
import networkx as nx
#import total2viz
from jpype import *
try:
    startJVM(getDefaultJVMPath(), "-Djava.class.path=D:/NLP/hanlp/hanlp-1.7.4.jar;D:/NLP/hanlp",
             "-Xms1g",
             "-Xmx1g")
except Exception:
    pass

#HanLP = JClass('com.hankcs.hanlp.HanLP')
#
#CRFnewSegment = HanLP.newSegment("crf")
#source='共青团深圳市宝安区委员会办公'
#term_list = CRFnewSegment.seg(source)
##print(term_list)
#thing_1='深圳市罗湖区教育局勤工俭学基建办公室  '
#str_=['区','团','局']
#total_list=[]
#def find_firstchar(str1,str_1_list):
#    str_list=list(str1)
#    count=0
#    for each_char in str_list:
#        for str_1 in str_1_list:
#            try:
##                print(''.join(str_list[count:count+len(str_1)])+' '+str_1)
##                print(str_1)
#                if(count+len(str_1)<len(str1)):
#                    if (''.join(str_list[count:count+len(str_1)])==str_1):
#                       return count
#            except Exception:
#                continue
#        count=count+1
#    return False
#while(find_firstchar(thing_1,str_)):
#                total_list.append(thing_1[0:find_firstchar(thing_1,str_)+1])
##                                            print(find_firstchar(thing_1,thing_2))
#                thing_1=thing_1[find_firstchar(thing_1,str_)+1:]
#print(total_list)
#import networkx as nx
#G = nx.Graph()
#nx.neighbors
#G.add_node('root')
#G.add_node('root1')
#G.add_node('root2')
#G.nodes['root1']['name']='123'
#G.add_edge('root','root1',weight=1)
##for thing in G.neighbors('root'):
##    print(thing['name'])
#print(G.nodes())
#def add_node(G,name):
#    G.add_node('213')
#add_node(G,'123')
#print(G.nodes())
#print([thing for thing in d.values()])
##    print(thing)
#print(fun)
#print(total_list)
import random
nid=0
def get_child_by_level_name(G,node,level,name,nid1=False):
    for thing in G.neighbors(node):
        if(G.nodes[thing]['level']==level+1 and name==G.nodes[thing]['name']):
            return thing  
    if(nid1==False):
        global nid
        nid=nid+1
        nid2=nid+1
    else:
        nid2=nid1
    G.add_node(str(nid2))
#    print(level+1)
    G.nodes[str(nid2)]['level']=level+1
    G.nodes[str(nid2)]['name']=name
#    print(name+str(level+1))
    if(level==2):
        random.seed()
#        G.add_edge(node,str(nid2),weight=int((np.exp(-(level))-np.abs(np.exp(-(level))*(random.random()-0.5)))*1000)/1000)
        G.add_edge(node,str(nid2),weight=int((np.exp(-(level))-1.1*np.abs(np.exp(-(level))*(random.random()-0.5)))*1000)/1000)
#        print(np.abs(np.exp(-(level))*(random.random()-0.5)))
    else:
        G.add_edge(node,str(nid2),weight=int(np.exp(-(level))*1000)/1000)
#    if(level==2):
#        G.add_edge(node,str(nid2),weight=(np.exp(-(level))-np.abs(np.exp(-(level))*(random.random()-0.5))))
##        print(np.abs(np.exp(-(level))*(random.random()-0.5)))
#    else:
#        G.add_edge(node,str(nid2),weight=np.exp(-(level)))
    return str(nid2)

def best_map(L1, L2):
    # L1 should be the groundtruth labels and L2 should be the clustering labels we got
    Label1 = np.unique(L1)
    nClass1 = len(Label1)
    Label2 = np.unique(L2)
    nClass2 = len(Label2)
    nClass = np.maximum(nClass1, nClass2)
    G = np.zeros((nClass, nClass))
    for i in range(nClass1):
        ind_cla1 = (L1 == Label1[i]).astype(float)
        for j in range(nClass2):
            ind_cla2 = (L2 == Label2[j]).astype(float)
            G[i, j] = np.sum(ind_cla2 * ind_cla1)
    m = Munkres()
    index = m.compute(-G.T)
    index = np.array(index)
    c = index[:, 1]
    newL2 = np.zeros(L2.shape)
    for i in range(nClass2):
        newL2[L2 == Label2[i]] = int(Label1[c[i]])
    return newL2 


def creat_g_fromlist(list1,threshold):	
#    from fuzzywuzzy import fuzz
    G=nx.Graph()
    G.add_node('-1')
    G.nodes['-1']['level']=0
    G.nodes['-1']['name']='root'
    value_list1=[int(list(value.keys())[0]) for value in list1]
    value_list=[list(value.keys())[0] for value in list1]
    global nid
    nid=max(value_list1)
#    print(nid)
    for inde,person in enumerate(list1):
        nid1=int(list(person.keys())[0])
        ins=person.values()
        ins_list=list(ins)[0].split(' ')
        next_node='-1'
        for index,thing in enumerate(ins_list):
            level=index
            if(index==len(ins_list)-1):
                next_node=get_child_by_level_name(G,next_node,level,thing,nid1)
            else:
                next_node=get_child_by_level_name(G,next_node,level,thing)
#    for thing in G.nodes():
#        if(G.nodes[thing]['level']==1):
#            print(str(G.nodes[thing]['level'])+G.nodes[thing]['name']+str(inde))
    from fuzzywuzzy import fuzz
    for thing in G.nodes():
        for thing1 in G.nodes():
            if(thing!=thing1 and G.nodes[thing]['level']==1 and G.nodes[thing1]['level']==1):            
                tempScore = fuzz.partial_ratio(G.nodes[thing]['name'], G.nodes[thing1]['name'])
                if(tempScore<80):
                    G.add_edge(thing,thing1,weight=1)
                else:
                    G.add_edge(thing,thing1,weight=1)
#    G.remove_node('-1')
#            print(123)
#            print(123)
    num=0
    for thing in G.nodes():
        if(G.nodes[thing]['level']==1):
            num=num+1
    print(num)
#    G.remove_node('-1')
    G1=nx.Graph()
    for index,source_node in enumerate(value_list):
        for target_node in value_list[index:]:
#            print(nx.shortest_path_length(G,source=source_node,target=target_node,weight='weight'))
            G1.add_node(source_node)
            G1.add_node(target_node)
            if(nx.shortest_path_length(G,source=source_node,target=target_node,weight='weight')<threshold):#1.83 is a nice parameter
                G1.add_edge(source_node,target_node,weight=nx.shortest_path_length(G,source=source_node,target=target_node,weight='weight'))
#    print()
    return G1

def creat_fuzzy_g(list1,ratio):	
#    from fuzzywuzzy import fuzz
    from fuzzywuzzy import fuzz
    G1=nx.Graph()
    value_list=[list(value.keys())[0] for value in list1]
    ins_list=[list(value.values())[0].replace(' ','') for value in list1]
    for index,source_node in enumerate(value_list):
        for index1,target_node in enumerate(value_list[index+1:]):
            G1.add_node(source_node)
            G1.add_node(target_node)
            if(fuzz.partial_ratio(ins_list[index],ins_list[index1])>ratio and index!=index1):
                G1.add_edge(source_node,target_node)  
#                print(ins_list[index]+' '+ins_list[index1])
#    G.remove_node('-1')
    return G1      

#def mix_label(list1,key,ratio):
#    for thing in list1:
#        small_list=thing.split(' ')
#        if('局' in small_list[0]): 
#            if()
def case_condition(G,count,list1):
    from random import sample
    l = G.edges() 
    list2=[]
    list3=[]
    element=sample(l, 5)# 随机抽取5个元素
    for index,thing in enumerate(element):
#            print('抽样样本第 '+str(index)+'个边')
            list2.append(list1[thing[0]])
            list3.append(list1[thing[1]])
    print(' & '.join(list2))
    print(' & '.join(list3))
    return list2,list3
list_500=[{'300': '公安局 办公室 办事员'}, {'1': '公安局 办公室 科员'}, {'2': '公安局 办公室 调研科副科长'}, {'3': '公安局 办公室 调研主任科员'}, {'4': '公安局 指挥部 秘书处 督查科科长'}, {'5': '公安局 指挥部 秘书处 处长党委委员'}, {'6': '公安局 指挥部 调研处调研科主任科员'}, {'7': '公安局 指挥部 信访处处长'}, {'8': '公安局 指挥部 指挥处处长 党总支委员'}, {'9': '公安局 指挥部 指挥处处长 党委委员'}, {'10': '公安局 宝安分局 黄田派出所 所长'}, {'11': '公安局 宝安分局 黄田派出所 办事员科员'}, {'12': '公安局 宝安分局 福永派出所科员'}, {'13': '公安局 宝安分局 龙城派出所科员'}, {'14': '公安局 宝安分局 指挥处处长党委委员'}, {'15': '公安局 宝安分局 副局长党委委员'}, {'16': '公安局 宝安分局 第一局长、党委副书记'}, {'17': '公安局 罗湖分局 科员'}, {'18': '公安局 罗湖分局 洪湖派出所 见习'}, {'19': '公安局 罗湖分局 洪湖派出所 科员'}, {'20': '公安局 罗湖分局 洪湖派出所 副主任科员'}, {'21': '公安局 罗湖分局 布心派出所副主任科员'}, {'22': '公安局 罗湖分局 东湖派出所副主任科员'}, {'23': '公安局 罗湖分局 副处级警官'}, {'24': '公安局 南山分局 西丽派出所 见习'}, {'25': '公安局 南山分局 西丽派出所 科员'}, {'26': '公安局 南山分局 南头派出所 履职副所长'}, {'27': '公安局 南山分局 南头派出所 副所长'}, {'28': '公安局 南山分局 塘朗派出所 教导员'}, {'29': '公安局 南山分局 塘朗派出所 所长'}, {'30': '公安局 南山分局 秘书科科员'}, {'31': '公安局 南山分局 妈湾派出所科员'}, {'32': '公安局 南山分局 巡警大队政工科副科长'}, {'33': '公安局 南山分局 政治处副主任'}, {'34': '公安局 南山分局 指挥处处长党委委员'}, {'35': '公安局 南山分局 副局长党委委员'}, {'36': '公安局 南山分局 粤海派出所综合室 主任'}, {'37': '公安局 南山分局 粤海派出所综合室 副主任'}, {'38': '公安局 南山分局 法制科科员副主任科员'}, {'39': '公安局 南山分局 预审科副主任科员'}, {'40': '公安局 南山分局 内保科科员'}, {'41': '公安局 南山分局 经济犯罪侦查大队长'}, {'42': '公安局 三 处 干部'}, {'43': '公安局 三 处 法制科 临时负责人'}, {'44': '公安局 三 处 法制科 副主任科员'}, {'45': '公安局 三 处 治安科干部'}, {'46': '公安局 三 治安案件查处、治安处机动队科员法制科科员'}, {'47': '公安局 政治部 综合处警察权益保障科副 科长'}, {'48': '公安局 政治部 综合处警察权益保障科副 主任科员'}, {'49': '公安局 政治部 训练处教育科科长'}, {'50': '公安局 沙河派出所 治安队副队长'}, {'51': '公安局 沙河派出所 社区防范队队长'}, {'52': '公安局 交警局科研所科员'}, {'53': '公安局 管理处治安防范科科员副'}, {'54': '公安局 龙岗分局消防监督管理大队科员建审科'}, {'55': '公安局 治安管理处 危管科副科长'}, {'56': '公安局 治安管理处 主任科员'}, {'57': '公安局 蛇口公安分局 四海派出所 见习'}, {'58': '公安局 蛇口公安分局 四海派出所 科员'}, {'59': '公安局 蛇口公安分局 预审科科员'}, {'60': '公安局 蛇口公安分局 法制科科员'}, {'61': '公安局 葵冲派出所教导员'}, {'62': '公安局 警察机动训练副支队'}, {'63': '公安局 南头派出所科员'}, {'64': '公安局 南高派出所副教导员'}, {'65': '公安局 桃源派出所副所长'}, {'66': '公安局 警务督察处副处长党总支委员'}, {'67': '公安局 二处 经保科未定级公务员'}, {'68': '公安局 二处 经济罪案调查一科员'}, {'69': '公安局 二处 政秘科副主任科员'}, {'70': '公安局 经济犯罪侦查支队 一大队 一 副中队长'}, {'71': '公安局 经济犯罪侦查支队 一大队 一 中队长'}, {'72': '公安局 经济犯罪侦查支队 一大队 副大队长'}, {'73': '公安局 经济犯罪侦查支队 情报大队长'}, {'74': '公安局 南湖派出所所长'}, {'75': '公安局 桂园派出所社区防范队队长'}, {'76': '公安局 东晓派出所教导员'}, {'77': '公安局 福田分局 刑警大队 三中队 副中队长'}, {'78': '公安局 福田分局 刑警大队 三中队 中队长'}, {'79': '公安局 福田分局 刑警大队 见习科员'}, {'80': '公安局 福田分局 刑警大队 副大队长'}, {'81': '公安局 福田分局 刑警大队 大队长'}, {'82': '公安局 福田分局 经侦大队 履行大队长职责'}, {'83': '公安局 福田分局 经侦大队 大队长'}, {'84': '公安局 福田分局 莲花派出所案件侦查队队长'}, {'85': '公安局 龙华分局政委、党委副书记'}, {'86': '公安局 福田分局梅林派出所刑警队队长'}, {'87': '光明新区 公明党工委 委员 组织人事办主任'}, {'88': '光明新区 公明党工委 委员 副书记'}, {'89': '光明新区 公明党工委 委员 书记兼办事处副主'}, {'90': '光明新区 公明党工委 书记 、办事处主任'}, {'91': '光明新区 公明党工委 副书记兼组织人事办主任'}, {'92': '光明新区 光明党工委 办公室主任'}, {'93': '光明新区 光明党工委 副书记、办事处主任'}, {'94': '光明新区 光明党工委 书记、办事处主任'}, {'95': '光明新区 人力资源办组织科科长'}, {'96': '光明新区 人事局组织科科长'}, {'97': '光明新区 社会管理局副局长'}, {'98': '光明新区 社会建设局副局长'}, {'99': '光明新区 光明办事处党政办主任'}, {'100': '光明新区 维稳一处处长'}, {'101': '光明新区 城市建设投资有限公司董事长'}, {'102': '光明新区 综合办公室副主任'}, {'103': '光明新区 安全生产监督管理局局长'}, {'104': '深圳能源集团 股份有限公司 企业文化部 党务'}, {'105': '深圳能源集团 股份有限公司 企业文化部 职总'}, {'106': '深圳能源集团 股份有限公司 企业文化部 总监'}, {'107': '深圳能源集团 股份有限公司 企业文化部 总经理'}, {'108': '深圳能源集团 股份有限公司 党委工作部部长'}, {'109': '深圳能源集团 股份有限公司 行政管理部总经理'}, {'110': '深圳能源集团 股份有限公司 管理部燃料运营计划高'}, {'111': '深圳能源集团 股份有限公司 产权法律部 总监'}, {'112': '深圳能源集团 股份有限公司 产权法律部 总经理'}, {'113': '深圳能源集团 股份有限公司 产权法律部 代职总监'}, {'114': '深圳能源集团 股份有限公司 规划发展部常规能源高级经'}, {'115': '深圳能源集团 企业文化部党支部支部书记'}, {'116': '深圳能源集团 党委工作部党支部书记'}, {'117': '深圳能源集团 计划发展部副主任'}, {'118': '深圳能源集团 规划发展部 副部长'}, {'119': '深圳能源集团 规划发展部 计划与统计高级经理'}, {'120': '深圳能源集团 发电分公司运行值长'}, {'121': '人民检察院 政治处 科员'}, {'122': '人民检察院 政治处 副主任科员'}, {'123': '人民检察院 侦查 三处主任科员'}, {'124': '人民检察院 侦查 监督部部长主任检察官检察'}, {'125': '人民检察院 侦查 一处 科员'}, {'126': '人民检察院 侦查 一处 主任科员'}, {'127': '人民检察院 侦查 一处副科级助检员'}, {'128': '人民检察院 侦查 监督科科长检委会委员'}, {'129': '人民检察院 民事行 政检察处正科级检察院'}, {'130': '人民检察院 民事行 政检察处副处长'}, {'131': '人民检察院 检察员'}, {'132': '人民检察院 科员试用期'}, {'133': '人民检察院 政治部主任科员'}, {'134': '人民检察院 组织教育处 处长'}, {'135': '人民检察院 组织教育处 副处长'}, {'136': '人民检察院 公共关系处处长'}, {'137': '人民检察院 诉讼监督部主任检察官'}, {'138': '人民检察院 党组成员 、反贪污贿赂局局长'}, {'139': '人民检察院 党组成员 反贪污贿赂局局'}, {'140': '人民检察院 副检察长党组成员、检察委'}, {'141': '人民检察院 干部助理检察员'}, {'142': '人民检察院 反贪污贿赂侦查 二科副科长'}, {'143': '人民检察院 反贪污贿赂侦查 一科副科长'}, {'144': '人民检察院 渎职侵权检察科科长'}, {'145': '人民检察院 反贪污贿赂局副局长'}, {'146': '人民检察院 光明检察室主任'}, {'147': '人民检察院 刑检一科科长检察员、检委'}, {'148': '人民检察院 公诉科科长检委会委员'}, {'149': '妈湾发电总厂 行政部 文秘科科长'}, {'150': '妈湾发电总厂 行政部 副部长'}, {'151': '妈湾发电总厂 行政部 代职部长'}, {'152': '妈湾发电总厂 运行 值长'}, {'153': '妈湾发电总厂 运行 部长'}, {'154': '妈湾发电总厂 运行 总监'}, {'155': '妈湾发电总厂 运行 部 单元长'}, {'156': '妈湾发电总厂 运行 部 值长'}, {'157': '妈湾发电总厂 运行 部 副值长'}, {'158': '妈湾发电总厂 综合部秘书'}, {'159': '妈湾发电总厂 党群部主持工作副部长'}, {'160': '妈湾发电总厂 安监部副部长'}, {'161': '妈湾发电总厂 生产准备部副部长'}, {'162': '妈湾发电总厂 脱硫工程部 负责人'}, {'163': '妈湾发电总厂 脱硫工程部 部长'}, {'164': '妈湾发电总厂 检修部部长'}, {'165': '妈湾发电总厂 检修总监'}, {'166': '住房和建设局 国际企业股份有限公司董事会秘书兼投资部经理'}, {'167': '住房和建设局 深飞激光光学系统有限责任公司生产主管'}, {'168': '住房和建设局 本守达进出口公司业务员'}, {'169': '住房和建设局 市民润农产品配送连锁商业有限公司筹备工作领导'}, {'170': '住房和建设局 是交通局运政监督分局机场交管站长'}, {'171': '住房和建设局 第26届世界大学生夏季运动会组委会执行局后勤保'}, {'172': '住房和建设局 公交集团公汽 技术部统计分析及技术开发工程师'}, {'173': '住房和建设局 公交集团公汽 生产部技术部长'}, {'174': '住房和建设局 公交集团公汽 三生产部副经理'}, {'175': '住房和建设局 巴士集团公共汽车分公司副总经理'}, {'176': '住房和建设局 市坪山区 环境保护和水务局调研员'}, {'177': '住房和建设局 市坪山区 城市更新局党组书记'}, {'178': '住房和建设局 商报社记者编辑'}, {'179': '住房和建设局 劳动时报编辑'}, {'180': '住房和建设局 警备区副参谋长'}, {'181': '政府 办公厅 第二秘书处 办事员'}, {'182': '政府 办公厅 第二秘书处 副处长'}, {'183': '政府 办公厅 综合处主任科员'}, {'184': '政府 办公厅 第一秘书处 主任科员'}, {'185': '政府 办公厅 第一秘书处 副处长'}, {'186': '政府 办公厅 社会处副处长'}, {'187': '政府 办公厅 副主任科员'}, {'188': '政府 办公厅 主任科员'}, {'189': '政府 办公厅 副处级秘书'}, {'190': '政府 办公厅 正处级秘书'}, {'191': '政府 应急管理办公室 预案 综合处副处长'}, {'192': '政府 应急管理办公室 预案 综合处调研员'}, {'193': '政府 应急管理办公室 市政府总值班室调研员'}, {'194': '政府 应急管理办公室 安全监察处调研员'}, {'195': '政府 应急管理办公室 应急指挥处 调研员'}, {'196': '政府 应急管理办公室 应急指挥处 处长'}, {'197': '政府 口岸办公室秘书处 科员'}, {'198': '政府 口岸办公室秘书处 副 处长'}, {'199': '政府 口岸办公室秘书处 副 主任科员'}, {'200': '政府 口岸办公室秘书处 主任科员'}, {'201': '政府 投资项目审计中心计财处副处长'}, {'202': '政府 办公室秘书科科副主任科员'}, {'203': '政府 法制办行政法规处副处长'}, {'204': '特区 建设发展集团有限公司 建设管理部 土建管理'}, {'205': '特区 建设发展集团有限公司 建设管理部 安装工程'}, {'206': '特区 建设发展集团有限公司 科技馆和创新大厦项'}, {'207': '特区 建设发展集团有限公司 巴新项目工作组负责'}, {'208': '特区 建设发展集团有限公司 国际低碳城项目部财'}, {'209': '特区 建设发展集团有限公司 规划发展部 项目发展'}, {'210': '特区 建设发展集团有限公司 规划发展部 副部长'}, {'211': '特区 建设发展集团有限公司 大铲湾滨海 新'}, {'212': '特区 建设发展集团有限公司 大铲湾滨海 新城项'}, {'213': '特区 建设发展集团有限公司 光明项目部总监'}, {'214': '特区 建设发展集团有限公司 人力资源部副部长'}, {'215': '特区 建设发展集团有限公司 党群工作部副部长'}, {'216': '特区 建设发展集团有限公司 太平洋工业区项目部'}, {'217': '特区 建设发展集团有限公司 平湖项目部项目总监'}, {'218': '特区 建设发展集团有限公司 深汕特别合作区项目'}, {'219': '特区 建发投资发展有限公司副总经理兼深圳市特'}, {'220': '特区 建发集团有限公司规划发展部预算岗'}, {'221': '人民政府 办公厅 信访办 综合处 处长'}, {'222': '人民政府 办公厅 信访办 综合处 主任科员'}, {'223': '人民政府 办公厅 信访办 综合处 副处长'}, {'224': '人民政府 办公厅 信访办 副主任科员'}, {'225': '人民政府 办公厅 信访办 办理处处长'}, {'226': '人民政府 办公厅 驻武汉办事处 试用期公务员'}, {'227': '人民政府 办公厅 驻武汉办事处 副主任科员'}, {'228': '人民政府 办公厅 第二秘书处 副 处长'}, {'229': '人民政府 办公厅 第二秘书处 副 主任科员'}, {'230': '人民政府 办公厅 第二秘书处 主任科员'}, {'231': '人民政府 办公厅 城建处主任科员'}, {'232': '人民政府 办公厅 社会一处副处长'}, {'233': '人民政府 外事办公室 秘书处 处长'}, {'234': '人民政府 外事办公室 秘书处 副处长'}, {'235': '人民政府 外事办公室 国际语言环境建设处处长'}, {'236': '人民政府 外事办公室 国际化促进处处长'}, {'237': '人民政府 外事办公室 港澳合作处处长'}, {'238': '人民政府 应急管理办公室 秘书处 调研员'}, {'239': '人民政府 应急管理办公室 秘书处 副处长'}, {'240': '人民政府 应急管理办公室 主任科员'}, {'241': '人民政府 文化产业发展办公室副主任'}, {'242': '人民政府 文化产业发展办公室综合协调处处长'}, {'243': '纪委 书记 ，区监委副主任人选'}, {'244': '纪委 书记 、区监委副主任人'}, {'245': '纪委 检查室 科员'}, {'246': '纪委 检查室 副 主任'}, {'247': '纪委 检查室 副 主任科员'}, {'248': '纪委 检查室 主任科员'}, {'249': '纪委 调研教育室科员'}, {'250': '纪委 副主任科员'}, {'251': '纪委 主任科员'}, {'252': '纪委 办公室副主任'}, {'253': '纪委 信访室主任'}, {'254': '纪委 监督室主任'}, {'255': '纪委 案件审理室主任'}, {'256': '纪委 常委案件审理室主任'}, {'257': '纪委 预防研究室主任科员'}, {'258': '农产品 股份有限公司 运通 事业部业务 一部长'}, {'259': '农产品 股份有限公司 运通 事业部业务 二部部长'}, {'260': '农产品 股份有限公司 运通 一部长'}, {'261': '农产品 股份有限公司 董事会办公室职员'}, {'262': '农产品 股份有限公司 资产经营 总部 副总经理'}, {'263': '农产品 股份有限公司 资产经营 总部 总经理兼市'}, {'264': '农产品 股份有限公司 资产经营 管理总部高级经理'}, {'265': '农产品 股份有限公司 历任总经理办公室秘书科副'}, {'266': '农产品 股份有限公司 市场投资管理总部副总经理'}, {'267': '农产品 股份有限公司 驻成都办事处主任'}, {'268': '农产品 股份有限公司 客服中心主任'}, {'269': '农产品 股份有限公司 电商总部总经理'}, {'270': '农产品 股份有限公司 企业发展部职员'}, {'271': '农产品 股份有限公司 总经理办公室 职员'}, {'272': '农产品 股份有限公司 总经理办公室 副 主任'}, {'273': '农产品 股份有限公司 总经理办公室 副 主任科员'}, {'274': '农产品 股份有限公司 总经理办公室 主任 助理'}, {'275': '农产品 批发公司公共关系部统计员总经理办公'}, {'276': '农产品 交易大厦有限公司总经理兼丰乐园大酒店'}, {'277': '农产品 集团股份有限公司 规划发展总部总经理'}, {'278': '农产品 集团股份有限公司 运通三部总经理'}, {'279': '深圳特区 报社 人事处 科员'}, {'280': '深圳特区 报社 人事处 干部调配科科长'}, {'281': '深圳特区 报社 政文部记者'}, {'282': '深圳特区 报社 校对科校对员'}, {'283': '深圳特区 报社 时事部编辑'}, {'284': '深圳特区 报业集团 社委办公室 法律事务部主管'}, {'285': '深圳特区 报业集团 社委办公室 秘书部主管'}, {'286': '深圳特区 报业集团 广告中心主任助理'}, {'287': '深圳特区 报业集团 对外新闻部 编辑'}, {'288': '深圳特区 报业集团 对外新闻部 助理调研员'}, {'289': '深圳特区 报 广告部 总经理'}, {'290': '深圳特区 报 广告部 常务副总经理'}, {'291': '深圳特区 报 新 媒体部 负责人'}, {'292': '深圳特区 报 新 媒体部 主任'}, {'293': '深圳特区 报 新 媒体运营部主任'}, {'294': '深圳特区 报 港澳台新闻部编辑'}, {'295': '深圳特区 报 香港记者站记者'}, {'296': '深圳特区 报 记者部记者'}, {'297': '深圳特区 报 国内新闻部编辑'}, {'298': '深圳特区 报 中国新闻部主任'}, {'299': '深圳特区 报社人事处干部调配科副科长'}, {'300': '深圳特区 报报业集团人力资源开发中心调配部主管'}, {'301': '深圳特区 报编委办公室 秘书'}, {'302': '深圳特区 报编委办公室 主任科员副科长'}, {'303': '文化局 编辑'}, {'304': '文化局 新闻出版处 科员'}, {'305': '文化局 新闻出版处 主任科员'}, {'306': '文化局 新闻出版处 副调研员'}, {'307': '文化局 社会文化处 科员'}, {'308': '文化局 社会文化处 主任科员'}, {'309': '文化局 社会文化处 副调研员'}, {'310': '文化局 广播电影电视管理处 科员'}, {'311': '文化局 广播电影电视管理处 副主任科员'}, {'312': '文化局 广播电影电视管理处 主任科员'}, {'313': '人事局 办公室 副主任'}, {'314': '人事局 办公室 主任科员'}, {'315': '人事局 调配处科员'}, {'316': '人事局 调配处副主任科员'}, {'317': '人事局 计划处主任科员'}, {'318': '人事局 人力资源开发处任主任科员'}, {'319': '人事局 人事编制信息处处长'}, {'320': '人事局 工资编制科科员副主'}, {'321': '人事局 公务员与职员 管理科科'}, {'322': '人事局 公务员与职员 管理处副处长'}, {'323': '人事局 专业技术人员管理科办事员科员'}, {'324': '人事局 录用考核任免科副科长主任科员'}, {'325': '人事局 教育培训科科长'}, {'326': '人事局 政策法规处'}, {'327': '中级人民法院 科员'}, {'328': '中级人民法院 督导室 副主任科员'}, {'329': '中级人民法院 督导室 主任科员'}, {'330': '中级人民法院 政治部 干部处主任科员'}, {'331': '中级人民法院 政治部 干部副处长'}, {'332': '中级人民法院 研究室 副 主任科 员'}, {'333': '中级人民法院 研究室 副 主任科 助理审判员'}, {'334': '中级人民法院 研究室 副 主任审判员'}, {'335': '中级人民法院 研究室 主任科 审判员'}, {'336': '中级人民法院 研究室 主任科 助理审判员'}, {'337': '中级人民法院 审判管理办公室 副主任审判员'}, {'338': '中级人民法院 审判管理办公室 主任审判员四级'}, {'339': '中级人民法院 法警支队副科级法警'}, {'340': '中级人民法院 助理审判员'}, {'341': '中级人民法院 房产庭副科级助理审判员'}, {'342': '中级人民法院 立案庭副科级助理审判员'}, {'343': '中级人民法院 立案庭正科级助理审判员'}, {'344': '中级人民法院 民事第五庭正科级审判员'}, {'345': '中级人民法院 房地产审判庭正科级审判员'}, {'346': '中级人民法院 办公室副主任审判员'}, {'347': '中级人民法院 司法警察支队副支队长'}, {'348': '教育局 党组 成员副局长'}, {'349': '教育局 党组 副书记区政府教育督导室主'}, {'350': '教育局 勤工俭学基建办公室科员'}, {'351': '教育局 行政办公室科员副主任'}, {'352': '教育局 成教与社会办学科副科长'}, {'353': '教育局 教育科科长'}, {'354': '教育局 勤工俭学基建科科长'}, {'355': '教育局 人事科干部副科长'}, {'356': '教育局 副局长'}, {'357': '教育局 专职党委副兼纪委书记'}, {'358': '教育局 团委负责人、团区委常委'}, {'359': '深圳大学 行政学系团委书记政治辅导员'}, {'360': '深圳大学 校友会办公室主任'}, {'361': '深圳大学 校友联络部主任'}, {'362': '深圳大学 国际贸易大专'}, {'363': '深圳大学 统计专业学习'}, {'364': '深圳大学 家用电器专业大专'}, {'365': '深圳大学 贸易系国际金融贸易专业学生'}, {'366': '深圳大学 电子系辅导员'}, {'367': '深圳大学 信息工程学院 辅导员'}, {'368': '深圳大学 信息工程学院 党委书记'}, {'369': '深圳大学 财务会计专业学习'}, {'370': '司法局 调研员'}, {'371': '司法局 办公室副主任'}, {'372': '司法局 社区矫正和安置帮教工作处 处长'}, {'373': '司法局 社区矫正和安置帮教工作处 副处长'}, {'374': '司法局 公证管理处副主任科员'}, {'375': '司法局 公证管理处主任科员'}, {'376': '司法局 律师管理处副处长'}, {'377': '司法局 律师公证管理处副处长'}, {'378': '司法局 政策法规处副'}, {'379': '司法局 政策法规处处'}, {'380': '司法局 副局长党组成员'}, {'381': '武汉大学 助教'}, {'382': '武汉大学 讲师'}, {'383': '武汉大学 副教授'}, {'384': '武汉大学 专门史专业硕士研究生'}, {'385': '武汉大学 分校历史专业学习'}, {'386': '武汉大学 图书馆系本科研究生'}, {'387': '武汉大学 政治系学生'}, {'388': '武汉大学 法律学专业大学本科'}, {'389': '武汉大学 国际经济法专业硕士研究生'}, {'390': '运输局 运政管理科 科长'}, {'391': '运输局 运政管理科 副科长'}, {'392': '运输局 交管员'}, {'393': '运输局 福田分局 稽查科 副科长'}, {'394': '运输局 福田分局 稽查科 主任科员'}, {'395': '运输局 福田分局 安技科副主任科员'}, {'396': '运输局 交通稽查大队大队长'}, {'397': '运输局 办公室科员'}, {'398': '运输局 建设处录用公务员'}, {'399': '运输局 建设处副主任科员'}, {'400': '运输局 规划建设处主任科员'}, {'401': '运输局 征稽办科员'}, {'402': '运输局 主任科员'}, {'403': '中山大学 电子与信息系统专业学生'}, {'404': '中山大学 历史学专业学习'}, {'405': '中山大学 中国近现代史专业硕士研究生'}, {'406': '中山大学 经济地理与城乡区域规划专业学习'}, {'407': '中山大学 经济学专业学生'}, {'408': '中山大学 经济法专业学生'}, {'409': '中山大学 自然地理专业学生'}, {'410': '中山大学 汉语言文学专业学生'}, {'411': '深圳水务集团 南山分公司管网部副部长'}, {'412': '深圳水务集团 管网部高级工程师'}, {'413': '深圳水务集团 运营部排水管网业务经理'}, {'414': '深圳水务集团 管网运营部副部长'}, {'415': '深圳水务集团 总工室副主任'}, {'416': '深圳水务集团 技术部部长'}, {'417': '深圳水务集团 机关五支部书记'}, {'418': '深圳水务集团 生产运营部部长'}, {'419': '水务集团 罗芳污水处理厂 党支部书记'}, {'420': '水务集团 罗芳污水处理厂 副厂长'}, {'421': '水务集团 维修中心副主任'}, {'422': '水务集团 东湖水厂副厂长'}, {'423': '水务集团 沙头角水厂副厂长'}, {'424': '水务集团 总工程师办公室工程师'}, {'425': '水务集团 信息中心副主任'}, {'426': '水务集团 客服分公司副经理'}, {'427': '水务集团 水表计量检定中心主任'}, {'428': '交通运输委 员会 西部交通运输局副局长'}, {'429': '交通运输委 员会 建设管理处 调研员'}, {'430': '交通运输委 员会 建设管理处 处长'}, {'431': '交通运输委 员会 物流发展处处长'}, {'432': '交通运输委 员会 综合运输处处长'}, {'433': '交通运输委 员会 光明交通运输局局长'}, {'434': '交通运输委 员会港航管理处副处长'}, {'435': '交通运输委 员会福田交通运输局副局长'}, {'436': '交通运输委 员会建设管理处副处长'}, {'437': '交通运输委 员会财务审计处副处长'}, {'438': '社保局 宝安分局 科员'}, {'439': '社保局 宝安分局 办公室 主任'}, {'440': '社保局 宝安分局 办公室 副主任'}, {'441': '社保局 宝安分局 副局长'}, {'442': '社保局 办公室主任科员'}, {'443': '社保局 医疗保险处主任科员'}, {'444': '社保局 机关事业社会保险处副处长'}, {'445': '社保局 计划财务处主任科员'}, {'446': '社保局 审计监督处副处长'}, {'447': '社保局 计划财务处副处长'}, {'448': '机场 股份有限公司 计划财务部副部长'}, {'449': '机场 股份有限公司 招标采购中心总经理'}, {'450': '机场 股份有限公司 党群工作部新闻中心业务经理'}, {'451': '机场 股份有限公司 机场新闻中心 对外宣传科外部'}, {'452': '机场 股份有限公司 机场新闻中心 副主任'}, {'453': '机场 有限公司 计划财务部副总经理'}, {'454': '机场 有限公司 商贸发展公司总经理'}, {'455': '机场 有限公司 审计法律部总经理市机'}, {'456': '机场 有限公司 党群工作部副总经理'}, {'457': '机场 市机场股份有限公司审计法律部副总经理'}, {'458': '光明 新区 人力资源办群团科主任科员'}, {'459': '光明 新区 公明工作委员会历任委员副书记'}, {'460': '光明 新区 规划土地监察局任局长'}, {'461': '光明 新区 发展和财政局副局长'}, {'462': '光明 畜牧场城管办干事'}, {'463': '光明 华侨畜牧场园林绿化队队长'}, {'464': '光明 农场人事科工作'}, {'465': '龙华新区 民治办事处党工委委员副主任'}, {'466': '龙华新区 龙华办事处 调研员'}, {'467': '龙华新区 龙华办事处 执法队队长'}, {'468': '龙华新区 龙华办事处 副处级领导干部'}, {'469': '龙华新区 观澜党工委委员宣传文化科科长'}, {'470': '龙华新区 观湖党工委委员、办事处副主任'}, {'471': '龙华新区 城市建设局局长'}, {'472': '监察局 办公室 主任科员'}, {'473': '监察局 办公室 副处级监察员'}, {'474': '监察局 预防 腐败室副主任'}, {'475': '监察局 预防 和研究室副主任'}, {'476': '监察局 信访举报室科员'}, {'477': '监察局 案件审理室副主任科员'}, {'478': '监察局 廉政教育处 科员'}, {'479': '监察局 廉政教育处 主任科员'}, {'480': '监察局 廉政教育处副主任科员'}, {'481': '监察局 政策法规室主任科员'}, {'482': '监察局 宣教和教研室主任科员'}, {'483': '监察局 法规和研究室副主任'}, {'484': '集团 军 124师372团 机枪连 排长'}, {'485': '集团 军 124师372团 机枪连 副指导员'}, {'486': '集团 军 124师372团 连指导员'}, {'487': '集团 军 团警侦连 战士'}, {'488': '集团 军 团警侦连 班长'}, {'489': '集团 军 124师372团政治处保卫股股长'}, {'490': '集团 军 124师政治部保卫科副营职干事'}, {'491': '集团 军 炮兵第一师 二十七 团三营九连排排长二营'}, {'492': '集团 军 炮兵第一师 二十七 团司令部作训股长'}, {'493': '集团 军 炮兵第一师 司令部 作训科营职参谋副科长'}, {'494': '集团 军 炮兵第一师 司令部 作训科科长'}, {'495': '集团 军 炮兵第一师 副参谋长'}, {'496': '集团 军 炮兵第一师 特务连指导员'}, {'497': '集团 军 炮兵第一师 二十六团代理参谋长'}, {'498': '集团 助理副主任维稳办公室专'}, {'499': '集团 调研员维稳办公室常务副主任'}]   
#list1=[{'200': '公安局 南山分局 政治处副主任'}, {'1': '公安局 南山分局 副局长党委委员'}, {'2': '公安局 南山分局 南头派出所履职副所长'}, {'3': '公安局 南山分局 粤海派出所综合室主任'}, {'4': '公安局 政治部 综合处警察权益保障科副主任科员'}, {'5': '公安局 政治部 训练处教育科科长'}, {'6': '公安局 宝安分局 副局长党委委员'}, {'7': '公安局 宝安分局 第一局长、党委副书记'}, {'8': '公安局 办公室办事员'}, {'9': '公安局 蛇口公安分局四海派出所见习'}, {'10': '公安局 葵冲派出所教导员'}, {'11': '公安局 经济犯罪侦查支队一大队一副中队长'}, {'12': '公安局 罗湖分局洪湖派出所 见习'}, {'13': '公安局 罗湖分局洪湖派出所 科员'}, {'14': '公安局 桂园派出所社区防范队队长'}, {'15': '公安局 东晓派出所教导员'}, {'16': '公安局 福田分局 经侦大队 履行大队长职责'}, {'17': '公安局 福田分局 经侦大队 大队长'}, {'18': '公安局 福田分局 刑警大队大队长'}, {'19': '公安局 福田分局 莲花派出所案件侦查队队长'}, {'20': '公安局 福田分局梅林派出所刑警队队长'}, {'21': '深圳特区 报社 政文部记者'}, {'22': '深圳特区 报社 时事部编辑'}, {'23': '深圳特区 报业集团 广告中心主任助理'}, {'24': '深圳特区 报业集团 对外新闻部 编辑'}, {'25': '深圳特区 报业集团 对外新闻部 助理调研员'}, {'26': '深圳特区 报 广告部总经理'}, {'27': '深圳特区 报 港澳台新闻部编辑'}, {'28': '深圳特区 报 记者部记者'}, {'29': '深圳特区 报 新媒体部负责人'}, {'30': '深圳特区 报 中国新闻部主任'}, {'31': '深圳特区 报社人事处干部调配科副科长'}, {'32': '深圳特区 报报业集团人力资源开发中心调配部主管'}, {'33': '深圳特区 报编委办公室 秘书'}, {'34': '深圳特区 报编委办公室 主任科员副科长'}, {'35': '深圳能源集团 股份有限公司 企业文化部总监'}, {'36': '深圳能源集团 股份有限公司 管理部燃料运营计划高'}, {'37': '深圳能源集团 股份有限公司 产权法律部总监'}, {'38': '深圳能源集团 股份有限公司 规划发展部常规能源高级经'}, {'39': '深圳能源集团 企业文化部党支部支部书记'}, {'40': '深圳能源集团 党委工作部党支部书记'}, {'41': '深圳能源集团 规划发展部计划与统计高级经理'}, {'42': '政府 办公厅 综合处主任科员'}, {'43': '政府 办公厅 第一秘书处主任科员'}, {'44': '政府 办公厅 第二秘书处副处长'}, {'45': '政府 办公厅 副主任科员'}, {'46': '政府 应急管理办公室 预案综合处副处长'}, {'47': '政府 应急管理办公室 应急指挥处调研员'}, {'48': '政府 口岸办公室秘书处 科员'}, {'49': '政府 口岸办公室秘书处 主任科员'}, {'50': '政府 投资项目审计中心计财处副处长'}, {'51': '政府 法制办行政法规处副处长'}, {'52': '运输局 交管员'}, {'53': '运输局 运政管理科科长'}, {'54': '运输局 办公室科员'}, {'55': '运输局 福田分局稽查科 副科长'}, {'56': '运输局 福田分局稽查科 主任科员'}, {'57': '运输局 建设处录用公务员'}, {'58': '运输局 规划建设处主任科员'}, {'59': '深圳大学 行政学系团委书记政治辅导员'}, {'60': '深圳大学 校友会办公室主任'}, {'61': '深圳大学 校友联络部主任'}, {'62': '深圳大学 贸易系国际金融贸易专业学生'}, {'63': '深圳大学 信息工程学院辅导员'}, {'64': '光明新区 人力资源办组织科科长'}, {'65': '光明新区 社会管理局副局长'}, {'66': '光明新区 城市管理局副局长'}, {'67': '光明新区 光明党工委书记、办事处主任'}, {'68': '光明新区 综合办公室副主任'}, {'69': '人民政府 办公厅 信访办 综合处主任科员'}, {'70': '人民政府 办公厅 信访办 办理处处长'}, {'71': '人民政府 办公厅 第二秘书处主任科员'}, {'72': '人民政府 办公厅 社会一处副处长'}, {'73': '人民政府 外事办公室国际化促进处处长'}, {'74': '人民政府 应急管理办公室秘书处副处长'}, {'75': '人民政府 文化产业发展办公室副主任'}, {'76': '人民检察院 政治处 科员'}, {'77': '人民检察院 政治处 副主任科员'}, {'78': '人民检察院 科员试用期'}, {'79': '人民检察院 侦查三处主任科员'}, {'80': '人民检察院 党组成员 、反贪污贿赂局局长'}, {'81': '人民检察院 党组成员 反贪污贿赂局局'}, {'82': '人民检察院 干部助理检察员'}, {'83': '人民检察院 反贪污贿赂侦查二科副科长'}, {'84': '人民检察院 反贪污贿赂局副局长'}, {'85': '妈湾发电总厂 运行部长'}, {'86': '妈湾发电总厂 生产准备部副部长'}, {'87': '妈湾发电总厂 脱硫工程部负责人'}, {'88': '妈湾发电总厂 检修部部长'}, {'89': '妈湾发电总厂 检修总监'}, {'90': '中山大学 电子与信息系统专业学生'}, {'91': '中山大学 历史学专业学习'}, {'92': '中山大学 经济学专业学生'}, {'93': '中山大学 经济法专业学生'}, {'94': '中学 教师'}, {'95': '纪委 检查室 科员'}, {'96': '纪委 检查室 副主任'}, {'97': '纪委 主任科员'}, {'98': '纪委 信访室主任'}, {'99': '纪委 案件审理室主任'}, {'100': '纪委 书记、区监委副主任人'}, {'101': '政法委 研究室副主任'}, {'102': '政法委 综合指导处副主任科员'}, {'103': '政法委 政策法规处副处长'}, {'104': '政法委 维稳一处处长'}, {'105': '水务集团 罗芳污水处理厂党支部书记'}, {'106': '水务集团 维修中心副主任'}, {'107': '水务集团 沙头角水厂副厂长'}, {'108': '水务集团 信息中心副主任'}, {'109': '深圳 是交通局运政监督分局机场交管站长'}, {'110': '深圳 巴士集团公共汽车分公司副总经理'}, {'111': '深圳 市坪山区环境保护和水务局调研员'}, {'112': '深圳 劳动时报编辑'}, {'113': '住房和建设局 财务审计处处长'}, {'114': '住房和建设局 勘察设计处处长'}, {'115': '住房和建设局 物业监管处处长'}, {'116': '农产品 股份有限公司 运通 事业部业务一部长'}, {'117': '农产品 股份有限公司 运通 一部长'}, {'118': '农产品 股份有限公司 资产经营总部副总经理'}, {'119': '农产品 股份有限公司 总经理办公室主任 助理'}, {'120': '农产品 批发公司公共关系部统计员总经理办公'}, {'121': '农产品 集团股份有限公司 规划发展总部总经理'}, {'122': '农产品 集团股份有限公司 运通三部总经理'}, {'123': '中国人民大学 法律系知识产权法第二学位班学生'}, {'124': '中国人民大学 区域经济学专业硕士研究生学习'}, {'125': '中国人民大学 档案系历史文献学专业学生'}, {'126': '教育局 党组 成员副局长'}, {'127': '教育局 党组 副书记区政府教育督导室主'}, {'128': '教育局 人事科干部副科长'}, {'129': '教育局 副局长'}, {'130': '教育局 专职党委副兼纪委书记'}, {'131': '广州铁路集团 公司长沙铁路总公司长沙机务段助理会计'}, {'132': '广州铁路集团 公司财务处出纳清算科会计稽查科等助'}, {'133': '广州铁路集团 公司资金结算中心计划部副经理会计师'}, {'134': '广东省 英德市团委书记'}, {'135': '广东省 沙头角林场干部'}, {'136': '广东省 茂名市人事局科员'}, {'137': '广东省英德市政府 党组成员市长助理'}, {'138': '广东省英德市政府 市长党组成员，市委政法委副书'}, {'139': '广东省英德市政府 副市长党组成员'}, {'140': '盐田港股份有限公司 港务部、操作部经理'}, {'141': '盐田港股份有限公司 安全管理委员会办公室副主'}, {'142': '盐田港股份有限公司 办公室主任'}, {'143': '龙华新区 民治办事处党工委委员副主任'}, {'144': '龙华新区 龙华办事处副处级领导干部'}, {'145': '龙华新区 观湖党工委委员、办事处副主任'}, {'146': '人事局 调配处副主任科员'}, {'147': '人事局 人事编制信息处处长'}, {'148': '人事局 政策法规处'}, {'149': '人大常委 会专职委员'}, {'150': '人大常委 会财政经济工作委员会主任'}, {'151': '人大常委 会龙岗街道工作委员会副主任'}, {'152': '人大常委 会办公室 副主任'}, {'153': '人大常委 会办公室 科员副主任科员'}, {'154': '中级人民法院 科员'}, {'155': '中级人民法院 研究室副主任科员'}, {'156': '中级人民法院 审判管理办公室副主任审判员'}, {'157': '社会保险基金管理局 医疗保险处副调研员'}, {'158': '社会保险基金管理局 宝安分局副调研员'}, {'159': '社会保险基金管理局 党委委员总会计师'}, {'160': '监察局 预防 腐败室副主任'}, {'161': '监察局 预防 和研究室副主任'}, {'162': '监察局 廉政教育处主任科员'}, {'163': '监察局 政策法规室主任科员'}, {'164': '监察局 宣教和教研室主任科员'}]
#list1=[{'18': '公安局 南山分局 政治处副主任'}, {'1': '公安局 南山分局 副局长党委委员'}, {'2': '公安局 南山分局 南头派出所履职副所长'}, {'3': '公安局 南山分局 粤海派出所综合室主任'}, {'4': '公安局 政治部 综合处警察权益保障科副主任科员'}, {'5': '公安局 政治部 训练处教育科科长'}, {'6': '公安局 宝安分局 副局长党委委员'}, {'7': '公安局 宝安分局 第一局长、党委副书记'}]
ColorList=['#e54d42','#f37b1d','#fbbd08', '#8dc63f', '#39b54a', '#1cbbb4','#0081ff','#6739b6', '#9c26b0','#e03997','#a5673f','#8799a3']
list1=list_500[0:200]
#print(len(list1))
dict1={}
value_list=[list(value.keys())[0] for value in list1]
index_list=[list(value.values())[0] for value in list1]
label_list=[list(value.values())[0].split(' ')[0] for value in list1]
from collections import Counter
num_Count=Counter(label_list)
#print(num_Count)
for index,thing in enumerate(value_list):
    dict1[thing]=index_list[index]
gt=label_list
list2 = gt
dic = {}
gt_short = dic.fromkeys(list2).keys()#
list3=[]
for thing in gt_short:
    list3.append(thing)
gt_short=list3
#gt_short=set(gt)
#print(gt_short)
gt_num=gt
for index,thing in enumerate(gt):
    for index1,thing1 in enumerate(gt_short):
        if(thing==thing1):
            gt_num[index]=index1
#print(gt_num)
#G1=creat_fuzzy_g(list1,30)
#print(gt_num)
#threshold=1.81
#while 1:
G1=creat_g_fromlist(list1,threshold)
G1=creat_fuzzy_g(list1,50)
G=G1
g=G


import os
import json
import copy
import networkx as nx
import random
import numpy as np
nodes={}
node_list=[]
node_flist=[]
link_list=[]
for year in range(1980,1981):
    import pandas as pd
    import json
    for node in G.nodes():
        h={}
        h['details']={}
        h['details']['age']=0
        h['details']['location']=0
        h['details']['organization']=0
        h['details']['position']=0
        h['details']['orgClass']=0
        h['id']=node
        h['type']=str(dict1[node]).split(' ')[0]
        h['year']=year
        node_flist.append(h)
    for link in G.edges():
        h={}
        h['source']=link[0]
        h['target']=link[1]
        h['weight']=0
        h['year']=year
        link_list.append(h)
        del h
print('init ready')
final={}
final['nodes']=node_flist
final['links']=link_list
with open('data.json', 'w') as f:
    json.dump(final, f,ensure_ascii=False)
    print('ok')
def conversion_coding(file_name, before='gbk', after='utf8'):
    # 此处转化文件编码，默认源文件编码为gbk，转换为utf8。
    # 原理： 读取文件中所有的内容（缓存到内存中），然后覆盖原文件（将缓存中的内容重新存入新文件）
    try:
        with open(file_name,'r',encoding = before ) as f:
            file_data = f.readlines()
        with open(file_name,'w',encoding = after) as f:
            f.writelines(file_data)
    except Exception as e:
        print('转换文件“{0}”失败！:{1}'.format(file_name, e))
conversion_coding('data.json')
```
